```python
import pandas as pd

data = pd.read_csv("./size.csv", header=None)
data
```

```python
dist = []
import numpy as np
import math
import seaborn as sns

for v1 in list(data[1]):
    temp = []
    for v2 in list(data[1]):
        temp.append(int(math.sqrt((int(v1)-int(v2))**2)))
    dist.append(temp)

for i, l in enumerate(dist):
    l[i]=9999999999
    
for i, l in enumerate(dist):
    print(i+1, " -> ", np.argmin(l) + 1, " : ", min(l))
```
