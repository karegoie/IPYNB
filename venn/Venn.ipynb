```python
import venn
import matplotlib.pyplot as plt
```

```python
braker = []
with open('./braker.anno', 'r') as f:
    for line in f:
        braker.append(line.strip())

interpro = []
with open('./interpro.anno', 'r') as f:
    for line in f:
        interpro.append(line.strip())

nr = []
with open('./nr.anno', 'r') as f:
    for line in f:
        nr.append(line.strip())

uniprot = []
with open('./uniprot.anno', 'r') as f:
    for line in f:
        uniprot.append(line.strip())
```

```python
labels = venn.get_labels([set(braker), set(interpro), set(nr)], fill=['number'])
venn.venn3(labels, names=["RNA-Seq, Ab inito based annotation(BRAKER2)", "Structure homology based(InterProScan)", "Sequence homology based(NR)"])
plt.show()
```

```python
labels = venn.get_labels([set(braker), set(interpro), set(nr), set(uniprot)], fill=['number'])
venn.venn3(labels, names=["RNA-Seq, Ab inito based annotation(BRAKER2)", "Structure homology based(InterProScan)", "Sequence homology based(NR)", "Sequence homology based(Uniprot)"])
plt.show()
```
