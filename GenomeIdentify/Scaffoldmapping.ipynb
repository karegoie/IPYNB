```python
from upsetplot import plot, from_memberships
import pandas as pd
import matplotlib.pyplot as plt
```

```python
scaff = {}
scaff_T = {}
with open('./temp1.txt', 'r') as f:
    for line in f:
        if not scaff.get(line.split('\t')[0].strip()): 
            scaff[line.split('\t')[0].strip()] = []
        if line.split('\t')[1].strip() not in scaff[line.split('\t')[0].strip()]: 
            scaff[line.split('\t')[0].strip()].append(line.split('\t')[1].strip())
        if not scaff_T.get(line.split('\t')[1].strip()):
            scaff_T[line.split('\t')[1].strip()] = []
        if line.split('\t')[0].strip() not in scaff_T[line.split('\t')[1].strip()]:
            scaff_T[line.split('\t')[1].strip()].append(line.split('\t')[0])
```

```python
scaffold = []
for k, v in sorted(scaff.items()):
    scaffold.append(v)
```

```python
scaffold = from_memberships(scaffold)
plt.figure(dpi=100000, figsize=(1000, 300))
from upsetplot import UpSet
ax_dict = UpSet(scaffold, subset_size='count', sort_by='cardinality').plot()
plt.savefig('./upsetplot.jpg')
```
