```python
import venn
import matplotlib.pyplot as plt
```

```python
M = []
with open('./missing_M.sorted.txt', 'r') as f:
    for line in f:
        if not line.startswith("#"):
            M.append(line.strip())

C = []
with open('./missing_C.sorted.txt', 'r') as f:
    for line in f:
        if not line.startswith("#"):
            C.append(line.strip())
```

```python
labels = venn.get_labels([set(M), set(C)], fill=['number'])
venn.venn3(labels, names=["MELVIO lineage", "CHAM lineage"])
plt.show()
```
