```python
from glob import glob
import numpy as np
filelist = glob("*.TPM.txt")
```

```python
import pandas as pd

data = {}
for name in filelist:
    data[name.split(".")[0]] = pd.read_csv(name, sep="\t", header=None)
```

```python
sorted_data = {}

for name in filelist:
    data[name.split(".")[0]].sort_values(0, inplace=True)
    sorted_data[name.split(".")[0]] = data[name.split(".")[0]][[0, 1]]
    
for k, v in sorted_data.items():
    v.columns=["gene", f"{k}"]
```

```python
sorted_data["vpF1-2"]
```

```python
for cnt, (k, v) in enumerate(sorted_data.items()):
    if cnt == 0:
        tmp = sorted_data[list(sorted_data.keys())[np.argmax(list(map(len, sorted_data.values())))]]
    else:
        if k != list(sorted_data.keys())[np.argmax(list(map(len, sorted_data.values())))]:
            tmp = pd.merge(tmp, sorted_data[k], on="gene", how="outer")

name = list(map(lambda x : x.split(".")[0], sorted(filelist)))
lost = list(set(name) - set(tmp.columns))[0]
tmp = pd.merge(tmp, sorted_data[lost], on="gene", how="outer")
name = list(map(lambda x : x.split(".")[0], sorted(filelist)))
tpmdata = tmp[["gene"] + name]
tpmdata
```

```python
tpmdata.to_csv("./TPMdata.csv", index=False)
```
